```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import acf
```

```python
data = pd.read_csv('data/commodity.txt', header=None, delimiter=',')
data.shape
```

```python
data.describe()
```

```python
data.head()
```

```python
acf_plot = plt.figure(figsize=(8,6))
plt.title('Commodity')
for i in range(10):
    acf_ndarray = acf(data[i].tolist(), nlags=50, fft=False)
    plt.plot(acf_ndarray)
acf_plot.show()
```
